````python
import IPython.display as display
import uuid
from IPython.display import HTML

def initialize_HTML_pages():
    zeichenfenster_ohne_ks = '''
    <!DOCTYPE html>
    <html lang="de">
    <head>
        <meta charset="UTF-8">
        <title>Zeichenfenster</title>
        <style>
            .modal {
                display: none;
                position: fixed;
                z-index: 1000;
                left: 0;
                top: 0;
                width: 100%;
                height: 100%;
                overflow: auto;
                background-color: rgba(0, 0, 0, 0.5);
            }
    
            .modal-content {
                background-color: white;
                margin: 5% auto;
                padding: 20px;
                border: 1px solid #888;
                width: 90%;
                max-width: 800px;
                display: flex;
                flex-direction: column;
                align-items: center;
            }
    
            #canvasContainer {
                width: 100%;
                padding-bottom: 75%; /* 4:3 Aspect Ratio */
                position: relative;
            }
    
            canvas {
                position: absolute;
                top: 0;
                left: 0;
                width: 100%;
                height: 100%;
                border: 1px solid #ccc;
            }
    
            #saveButton {
                margin-top: 10px;
            }
        </style>
    </head>
    <body>
        <button id="openModalButton">Zeichenfenster oeffnen</button>
        <div id="drawingModal" class="modal">
            <div class="modal-content">
                <span id="closeModalButton" style="cursor:pointer; align-self: flex-end;">&times;</span>
                <div id="canvasContainer">
                    <canvas id="drawingCanvas"></canvas>
                </div>
                <button id="saveButton">Zeichnung speichern</button>
            </div>
        </div>
        <script>
            const canvas = document.getElementById("drawingCanvas");
            const ctx = canvas.getContext("2d");
            let isDrawing = false;
            let startX, startY;
    
            function resizeCanvas() {
                const container = canvas.parentElement;
                canvas.width = container.offsetWidth;
                canvas.height = container.offsetHeight;
            }
    
            window.addEventListener("resize", resizeCanvas);
            resizeCanvas();
    
            canvas.addEventListener("mousedown", (e) => {
                isDrawing = true;
                [startX, startY] = [e.offsetX, e.offsetY];
            });
    
            canvas.addEventListener("mousemove", (e) => {
                if (!isDrawing) return;
                ctx.beginPath();
                ctx.moveTo(startX, startY);
                ctx.lineTo(e.offsetX, e.offsetY);
                ctx.strokeStyle = "red";
                ctx.lineWidth = 2;
                ctx.stroke();
                [startX, startY] = [e.offsetX, e.offsetY];
            });
    
            canvas.addEventListener("mouseup", () => {
                isDrawing = false;
            });
    
            document.getElementById("openModalButton").onclick = function () {
                document.getElementById("drawingModal").style.display = 'block';
                resizeCanvas();
            }
    
            document.getElementById("closeModalButton").onclick = function () {
                document.getElementById("drawingModal").style.display = 'none';
            }
    
            document.getElementById("saveButton").addEventListener("click", () => {
                const drawingData = canvas.toDataURL("image/png");
                fetch("/save_drawing", {
                    method: "POST",
                    headers: { "Content-Type": "application/json" },
                    body: JSON.stringify({ drawing: drawingData })
                })
                    .then(response => response.json())
                    .then(data => alert(data.message))
                    .catch(error => console.error("Fehler:", error));
            });
        </script>
    </body>
    </html>'''
    
    
    zeichenfenster = '''
    <!DOCTYPE html>
    <html lang="de">
    <head>
        <meta charset="UTF-8">
        <title>Zeichenfenster</title>
        <style>
            canvas {
                border: 1px solid black;
            }
            /* Modal Styles */
            .modal {
                display: none; /* Versteckt das Modal standardm√§√üig */
                position: fixed; /* Bleibt an der gleichen Stelle */
                z-index: 1000; /* √úber anderen Inhalten */
                left: 0;
                top: 0;
                width: 100%; /* Vollbild */
                height: 100%; /* Vollbild */
                overflow: auto; /* Scrollen falls n√∂tig */
                background-color: rgba(0, 0, 0, 0.5); /* Halbtransparentes Schwarz */
            }
    
            .modal-content {
                background-color: white;
                margin: 15% auto; /* Zentriert das Modal */
                padding: 20px;
                border: 1px solid #888;
                width: 80%; /* Breite des Modals */
            }
        </style>
    </head>
    <body>
    
        <!-- Button zum √ñffnen des Modals -->
        <button id="openModalButton">Zeichenfenster √∂ffnen</button>
    
        <!-- Das Modal -->
        <div id="drawingModal" class="modal">
            <div class="modal-content">
                <span id="closeModalButton" style="cursor:pointer;">&times;</span>
                <canvas id="drawingCanvas" width="800" height="600"></canvas>
                <button id="saveButton">Zeichnung speichern</button>
            </div>
        </div>
    
        <script src="https://cdnjs.cloudflare.com/ajax/libs/socket.io/4.5.4/socket.io.js"></script>
        <script>
        const canvas = document.getElementById('drawingCanvas');
        const ctx = canvas.getContext('2d');
        let isDrawing = false;
        let startX, startY;
    
        // Funktion zum Zeichnen des Koordinatensystems
        function drawCoordinateSystem() {
            const width = canvas.width;
            const height = canvas.height;
            const centerX = width / 2;
            const centerY = height / 2;
            const scale = 40;
    
            // X-Achse
            ctx.beginPath();
            ctx.moveTo(0, centerY);
            ctx.lineTo(width, centerY);
            ctx.stroke();
    
            // Y-Achse
            ctx.beginPath();
            ctx.moveTo(centerX, 0);
            ctx.lineTo(centerX, height);
            ctx.stroke();
    
            // Markierungen und Zahlen
            for (let i = -10; i <= 10; i++) {
                ctx.fillText(i, centerX + i * scale, centerY + 20);
                ctx.beginPath();
                ctx.moveTo(centerX + i * scale, centerY - 5);
                ctx.lineTo(centerX + i * scale, centerY + 5);
                ctx.stroke();
    
                ctx.fillText(i, centerX - 20, centerY - i * scale);
                ctx.beginPath();
                ctx.moveTo(centerX - 5, centerY - i * scale);
                ctx.lineTo(centerX + 5, centerY - i * scale);
                ctx.stroke();
            }
        }
    
        // Event Listener f√ºr das Zeichnen
        canvas.addEventListener('mousedown', (e) => {
            isDrawing = true;
            [startX, startY] = [e.offsetX, e.offsetY];
        });
    
        canvas.addEventListener('mousemove', (e) => {
            if (!isDrawing) return;
            ctx.beginPath();
            ctx.moveTo(startX, startY);
            ctx.lineTo(e.offsetX, e.offsetY);
            ctx.strokeStyle = 'red';
            ctx.lineWidth = 2;
            ctx.stroke();
            [startX, startY] = [e.offsetX, e.offsetY];
        });
    
        canvas.addEventListener('mouseup', () => {
            isDrawing = false;
        });
    
        drawCoordinateSystem();
    
        // Funktion zum √ñffnen des Modals
        document.getElementById('openModalButton').onclick = function() {
            document.getElementById('drawingModal').style.display = 'block';
        }
    
        // Funktion zum Schlie√üen des Modals
        document.getElementById('closeModalButton').onclick = function() {
            document.getElementById('drawingModal').style.display = 'none';
        }
    
        // Speichern der Zeichnung
        document.getElementById('saveButton').addEventListener('click', () => {
            const drawingData = canvas.toDataURL('image/png'); // Zeichnung als Base64-Daten
            fetch('/save_drawing', {
                method: 'POST',
                headers: { 'Content-Type': 'application/json' },
                body: JSON.stringify({ drawing: drawingData })
            })
            .then(response => response.json())
            .then(data => alert(data.message))
            .catch(error => console.error('Fehler:', error));
        });
        </script>
    
    </body>
    </html>'''
    
    
    yt_kommentaranalyse = '''
    <!DOCTYPE html>
    <html lang="de">
    <head>
        <meta charset="UTF-8">
        <title>YouTube Kommentar-Analyse</title>
        <style>
            body {
                font-family: Arial, sans-serif;
                max-width: 600px;
                margin: 0 auto;
                padding: 20px;
            }
    
            button {
                font-size: 16px;
                padding: 10px;
                margin: 10px 0;
            }
    
            textarea {
                width: 100%;
                padding: 10px;
                margin: 10px 0;
            }
    
            #result {
                border: 1px solid #ccc;
                padding: 10px;
                margin-top: 20px;
            }
        </style>
    </head>
    <body>
        <h1>YouTube Kommentar-Analyse</h1>
    
        <p>
            Wir werden im Folgenden die Emotionen eines YouTube-Kommentars auswerten.<br>
            √ñffne YouTube mit dem Button, suche ein Video, kopiere einen Kommentar und f√ºge ihn hier ein.
        </p>
    
        <button onclick="openYouTube()">YouTube-Suche √∂ffnen</button>
    
        <p>Hier den YouTube-Kommentar einf√ºgen:</p>
        <textarea id="commentInput" rows="2" cols="50"></textarea>
    
        <button onclick="analyzeText()">Auswerten</button>
        <button onclick="clearText()">Leeren</button>
    
        <h2>Ergebnis:</h2>
        <div id="result"></div>
    
        <script>
            function openYouTube() {
                window.open("https://www.youtube.com/results?search_query=", "_blank");
            }
    
            function analyzeText() {
                var comment = document.getElementById("commentInput").value;
                if (comment.trim() === "") {
                    alert("Bitte Text eingeben!");
                    return;
                }
                // Hier w√ºrde normalerweise die Analyse-Logik stehen
                document.getElementById("result").innerText = "Analyse-Ergebnis w√ºrde hier erscheinen.";
            }
    
            function clearText() {
                document.getElementById("commentInput").value = "";
                document.getElementById("result").innerText = "";
            }
        </script>
    </body>
    </html>'''
    
    
    kameraanalyse = '''
    <!DOCTYPE html>
    <html lang="de">
    <head>
        <meta charset="UTF-8">
        <title>Camera App</title>
        <style>
            body {
                font-family: Arial, sans-serif;
                text-align: center;
                margin-top: 50px;
            }
    
            #cameraWindow {
                display: none;
                margin-top: 20px;
            }
    
            button {
                font-size: 16px;
                padding: 10px;
                margin: 10px;
            }
        </style>
    </head>
    <body>
        <h1>Main Window</h1>
        <button onclick="openCameraWindow()">Open Camera Window</button>
    
        <div id="cameraWindow">
            <h2>Camera Window</h2>
            <video id="video" width="640" height="480" autoplay></video>
            <br>
            <button onclick="captureAndClassify()">Capture and classify image</button>
            <br>
            <canvas id="canvas" width="640" height="480" style="display:none;"></canvas>
        </div>
    
        <script>
            let stream;
    
            function openCameraWindow() {
                document.getElementById('cameraWindow').style.display = 'block';
                startVideoStream();
            }
    
            async function startVideoStream() {
                try {
                    stream = await navigator.mediaDevices.getUserMedia({ video: true });
                    document.getElementById('video').srcObject = stream;
                } catch (err) {
                    console.error("Error: " + err);
                    alert("Failed to initialize the camera.");
                }
            }
    
            function captureAndClassify() {
                const video = document.getElementById('video');
                const canvas = document.getElementById('canvas');
                const context = canvas.getContext('2d');
    
                // Capture the current video frame
                context.drawImage(video, 0, 0, canvas.width, canvas.height);
    
                // Convert the captured image to base64
                const imageDataUrl = canvas.toDataURL('image/jpeg');
    
                // Here you would normally send the image data to a server for classification
                // For this example, we'll just show a placeholder result
                classifyImage(imageDataUrl);
            }
    
            function classifyImage(imageData) {
                // Placeholder for image classification
                // In a real application, you would send this data to a server for processing
                alert("Image captured and ready for classification.\nClassification result would appear here.");
            }
    
            // Clean up resources when the page is closed
            window.onbeforeunload = function() {
                if (stream) {
                    stream.getTracks().forEach(track => track.stop());
                }
            };
        </script>
    </body>
    </html>'''
    
    
    
    audioaufnahme = '''
    <!DOCTYPE html>
    <html lang="de">
    <head>
        <meta charset="UTF-8">
        <title>Audio Aufnahme App</title>
        <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
        <style>
            #audioGraph {
                width: 100%;
                height: 300px;
            }
        </style>
    </head>
    <body>
        <button id="startRecording">Aufnahme starten</button>
        <button id="stopRecording" disabled>Aufnahme stoppen</button>
        <div id="audioGraph"></div>
    
        <script>
            let audioContext;
            let audioStream;
            let audioProcessor;
            let audioQueue = [];
            const QUEUE_SIZE = 50;
            let isRecording = false;
            const THRESHOLD = 0.01; // Schwellenwert f√ºr Audioaufnahme
    
            async function startRecording() {
                try {
                    audioContext = new (window.AudioContext || window.webkitAudioContext)();
                    audioStream = await navigator.mediaDevices.getUserMedia({ audio: true });
                    const source = audioContext.createMediaStreamSource(audioStream);
                    audioProcessor = audioContext.createScriptProcessor(4096, 1, 1);
    
                    source.connect(audioProcessor);
                    audioProcessor.connect(audioContext.destination);
    
                    audioProcessor.onaudioprocess = (e) => {
                        const inputData = e.inputBuffer.getChannelData(0);
                        processAudio(inputData);
                    };
    
                    isRecording = true;
                    document.getElementById('startRecording').disabled = true;
                    document.getElementById('stopRecording').disabled = false;
                } catch (err) {
                    console.error('Fehler beim Starten der Aufnahme:', err);
                }
            }
    
            function stopRecording() {
                if (audioStream) {
                    audioStream.getTracks().forEach(track => track.stop());
                }
                if (audioProcessor) {
                    audioProcessor.disconnect();
                }
                if (audioContext) {
                    audioContext.close();
                }
                isRecording = false;
                document.getElementById('startRecording').disabled = false;
                document.getElementById('stopRecording').disabled = true;
            }
    
            function processAudio(inputData) {
                const reducedData = inputData
                    .filter((_, index) => index % 20 === 0)
                    .map(value => {
                        const absValue = Math.abs(value);
                        if (absValue < THRESHOLD) {
                            return 0;
                        } else {
                            // Logarithmische Skalierung
                            const sign = value < 0 ? -1 : 1;
                            return sign * Math.log10(1 + 9 * absValue);
                        }
                    });
    
                audioQueue.push(...reducedData);
    
                if (audioQueue.length > QUEUE_SIZE) {
                    audioQueue = audioQueue.slice(-QUEUE_SIZE);
                }
    
                updateGraph();
                sendDataToServer();
            }
    
            function updateGraph() {
                const trace = {
                    y: audioQueue,
                    type: 'scatter',
                    mode: 'lines',
                    line: {
                        color: 'blue',
                        width: 2
                    }
                };
    
                const layout = {
                    title: 'Audio Waveform (Logarithmische Amplitude)',
                    xaxis: { title: 'Sample' },
                    yaxis: {
                        title: 'Logarithmische Amplitude',
                        range: [-1, 1] // Angepasster Bereich f√ºr logarithmische Skala
                    }
                };
    
                Plotly.newPlot('audioGraph', [trace], layout);
            }
    
            function sendDataToServer() {
                // Senden der Daten an das √ºbergeordnete Fenster (Dash)
                // window.parent.postMessage({ type: 'audioData', data: audioQueue }, '*');
                fetch('/update_audio', {
                    method: 'POST',
                    headers: {
                        'Content-Type': 'application/json',
                    },
                    body: JSON.stringify({ audioData: audioQueue }),
                })
                    .then(response => response.json())
                    .then(data => console.log('Server response:', data))
                    .catch((error) => console.error('Error:', error));
            }
    
            document.getElementById('startRecording').addEventListener('click', startRecording);
            document.getElementById('stopRecording').addEventListener('click', stopRecording);
        </script>
    </body>
    </html>'''

initialize_HTML_pages()
````
