Jetzt wird die Trainings-Pipeline definiert:

```python
from autotrain import AutoTrainPipeline

pipeline = AutoTrainPipeline(
    task="text_classification",
    model="bert-base-uncased",
    data_path="emotion_data.csv",
    column_mapping={"text": "text", "label": "label"},
    args={
        "lr": 2e-5,          # Lernrate
        "epochs": 3,         # Anzahl der Durchläufe durchs gesamte Dataset
        "batch_size": 8      # Wie viele Beispiele pro Trainingsschritt verarbeitet werden
    }
)
```

🧠 **Erklärung der wichtigsten Parameter:**

- `lr` (*learning rate*): Gibt an, wie stark das Modell bei jedem Schritt lernt. Kleinere Werte führen zu stabilerer, aber langsamerer Anpassung.
- `epochs`: Wie oft das gesamte Trainingsset durchlaufen wird. Mehr Epochen = mehr Lernzeit.
- `batch_size`: Anzahl der Texte, die pro Trainingsschritt gleichzeitig verarbeitet werden. Große Werte = schneller, aber auch speicherintensiver.
