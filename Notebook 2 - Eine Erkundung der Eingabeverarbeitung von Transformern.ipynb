```python
import ipywidgets as widgets
from IPython.display import display

# --- Funktion: Ein Punkt mit ausklappbarem Bereich + Bild rechts ---
def punkt_mit_ausklappbereich_und_bild(label, bild_url):
    toggle = widgets.ToggleButton(
        description=label,
        button_style='',
        layout=widgets.Layout(width="100%", margin="3px 0")
    )

    # Linker Textbereich
    text_links = widgets.HTML(
        f"<p><b>{label}</b>: Dies ist der eingeblendete Text mit etwas Erklaerung.</p>"
    )

    # Rechter Bildbereich
    bild_rechts = widgets.HTML(
        f'''
        <div style="width:100%; text-align:center;">
            <img src="{bild_url}" style="width:100%; height:auto;"/>
        </div>
        '''
    )

    # Zwei-Spalten-Layout für den Inhalt
    inhalt = widgets.GridBox(
        children=[text_links, bild_rechts],
        layout=widgets.Layout(
            grid_template_columns="62% 38%",
            grid_gap="10px",
            display="none",  # Anfangszustand: versteckt
            border="1px dotted #aaa",
            padding="6px"
        )
    )

    def show_hide(change):
        inhalt.layout.display = "grid" if change["new"] else "none"

    toggle.observe(show_hide, names="value")
    return widgets.VBox([toggle, inhalt])

# --- Funktion: Untergrid mit mehreren Punkten ---
def erstelle_untergrid(name):
    header = widgets.HTML(f"<h4>{name}</h4>")
    punkte = [
        punkt_mit_ausklappbereich_und_bild(f"{name} – Punkt {i}", "https://i.imgur.com/I4WbNhe.png")
        for i in range(1, 4)
    ]
    return widgets.VBox([header] + punkte, layout=widgets.Layout(
        border="1px solid #ccc", padding="6px", margin="5px 0"
    ))

# --- Linke Spalte ---
links = widgets.VBox([
    erstelle_untergrid("🧩 Teil 1"),
    erstelle_untergrid("🧪 Teil 2")
], layout=widgets.Layout(
    border="2px solid #aaa", padding="10px", width="100%"
))

# --- Rechte Spalte ---
rechts = widgets.HTML('''
    <div style="width: 100%;">
        <img src="https://i.imgur.com/TDUsRIF.png" style="width: 95%; height: auto;"/>
    </div>
''')

# --- Grid Layout ---
grid = widgets.GridBox(
    children=[links, rechts],
    layout=widgets.Layout(
        grid_template_columns="36% 60%",
        grid_gap="20px"
    )
)

# --- Anzeige ---
display(grid)
```

#### **Aufgabe 5** (das Skalarprodukt)

##### **Definition: Skalarprodukt**

Es seien zwei Vektoren gegeben:  

$$
\vec{a} = \begin{pmatrix} a_1 \\ a_2 \\ \dots \\ a_n \end{pmatrix}, \quad
\vec{b} = \begin{pmatrix} b_1 \\ b_2 \\ \dots \\ b_n \end{pmatrix}
$$

Es existieren zwei Formeln für das Skalarprodukt, welche wir beide in den folgenden Kapiteln benötigen werden.

**1. Algebraische Formel:**  
Man multipliziert die jeweiligen Komponenten $a_i$ mit den entsprechenden Komponenten $b_i$ und addiert die Ergebnisse.  
Die Formel lautet:  
$$
\vec{a} \cdot \vec{b} = \sum_{i=1}^n a_i b_i = a_1 b_1 + a_2 b_2 + \dots + a_n b_n
$$

<details>
<summary>💡 <b>Details & Beispiel</b> </summary>

Die algebraische Sicht bietet eine schnelle und effiziente Möglichkeit, das Skalarprodukt zu berechnen. Sie eignet sich besonders dann, wenn nur der reine Wert benötigt wird, ohne dass geometrische Interpretationen im Vordergrund stehen. Gerade in höheren Dimensionen ist diese Form praktisch, da sie sich direkt aus den Koordinaten der Vektoren ergibt und unkompliziert berechnet werden kann.

Folgende Operationen werden dabei ausgeführt:

- Jede Komponente $a_i$ wird mit der entsprechenden Komponente $b_i$ multipliziert.  
- Danach werden alle Produkte addiert.  

Beispiel: Seien 
$$
\vec{a} = \begin{pmatrix} 2.5 \\ 4 \end{pmatrix}, \quad
\vec{b} = \begin{pmatrix} 5 \\ 3 \end{pmatrix}
$$.

Dann lautet die Rechnung:

$$
\vec{a} \cdot \vec{b} = 2.5 \cdot 5 + 4 \cdot 3 = 12.5 + 12 = 24.5
$$

</details>

**2. Geometrische Formel:**  
Mit der folgenden Formel lässt sich das Skalarprodukt ebenfalls berechnen:  
$$
\vec{a} \cdot \vec{b} = |\vec{a}| \, |\vec{b}| \, \cos(\theta)
$$  
wobei $\theta$ der Winkel zwischen den Vektoren ist.

*Die Rechenvorschrift der geometrischen Formel ist folgende (im zweidimensionalen Raum):* 

Gegeben seien folgende Vektoren und folgender Winkel:

$$
\vec{a}=\begin{pmatrix}a_1\\a_2\end{pmatrix},\qquad
\vec{b}=\begin{pmatrix}b_1\\b_2\end{pmatrix},\qquad
\text{Winkel }\theta=\dots^\circ
$$

**1) Beträge berechnen**

$$
|\vec{a}|=\sqrt{a_1^2+a_2^2},\qquad
|\vec{b}|=\sqrt{b_1^2+b_2^2}
$$

**2) Kosinus des Winkels berechnen**

$$
\cos(\theta)=\dots
$$

**3) Skalarprodukt berechnen**

$$
\vec{a}\cdot\vec{b}=|\vec{a}|\,|\vec{b}|\,\cos(\theta)
$$

<details>
<summary>💡 <b>Details & Beispiel</b></summary>

Die geometrische Sicht verdeutlicht die inhaltliche Bedeutung des Skalarproduktes. Es verbindet die Länge zweier Vektoren mit dem Winkel, den sie einschließen, und macht dadurch ihre räumliche Ausrichtung sichtbar. Diese Sichtweise eignet sich besonders zur Winkelberechnung oder zum Orthogonalitätstest.

Folgende Parameter werden in der Formel verarbeitet:

- $|\vec{a}|$ und $|\vec{b}|$ sind die Beträge der beiden Vektoren – d. h. die Längen.  
- $\theta$ ist der eingeschlossene Winkel.


**Anhand eines konkreten Beispiels:**


Gegeben seien die Vektoren und der Winkel:

$$
\vec{a}=\begin{pmatrix}2.5\\4\end{pmatrix},\qquad
\vec{b}=\begin{pmatrix}5\\3\end{pmatrix},\qquad
\theta\approx 27.03^\circ
$$

<img src="https://i.imgur.com/qq6YAHf.jpeg" alt="Vektorgrafik" width="400">

**1) Beträge berechnen**

$$
|\vec{a}|=\sqrt{2.5^2+4^2}=\sqrt{22.25}\approx 4.72,\qquad
|\vec{b}|=\sqrt{5^2+3^2}=\sqrt{34}\approx 5.83
$$

**2) Kosinus des Winkels berechnen**

$$
\cos(\theta)=\cos(27.03^\circ)\approx 0.891
$$

**3) Skalarprodukt berechnen**

$$
\vec{a}\cdot\vec{b}=|\vec{a}|\,|\vec{b}|\,\cos(\theta)\approx 24.5
$$

</details>


---
**Folgende Eigenschaft des Skalarproduktes wird im Kapitel zur "Kosinus-Ähnlichkeit" für uns von Bedeutung sein:**  

- Ist das Skalarprodukt $= 0$, dann stehen die Vektoren **orthogonal** (senkrecht) zueinander.  
- Ist das Skalarprodukt $= +1$, dann zeigen die Vektoren in **dieselbe Richtung** (maximale Ausrichtung).  
- Ist das Skalarprodukt $= -1$, dann zeigen die Vektoren in **entgegengesetzte Richtungen**.

### **Block 5: Kosinusähnlichkeit und semantische Suche**

In diesem Abschnitt lernt ihr, wie man die **Bedeutungsähnlichkeit zwischen Wörtern bzw. Tokens** mathematisch bestimmen kann – mithilfe der **Kosinus-Ähnlichkeit**. Sie hilft uns zu erkennen, ob zwei *Worteinbettungen* (also mentale Repräsentationen von Wörtern) ähnlich oder gegensätzlich sind – genau das, was Modelle wie **Google Translate** oder **ChatGPT** ständig tun.  

$$
\cos(\theta) = \frac{\vec{a} \cdot \vec{b}}{|\vec{a}| \cdot |\vec{b}|}
$$

Wie ihr seht, handelt es sich bei den Embeddings um 768-dimensionale Vektoren. Um die Ähnlichkeit solcher hochdimensionalen Vektoren zu vergleichen, wird oft die **Kosinus-Ähnlichkeit** genutzt.

$$
\cos(\theta) = \frac{\vec{a} \cdot \vec{b}}{|\vec{a}| \cdot |\vec{b}|}
$$

Wie aus den Folien bekannt, gilt:  
- **+1** = sehr ähnlich (zeigen in gleiche Richtung),  
- **0** = keine Ähnlichkeit (stehen senkrecht),  
- **-1** = gegensätzlich (zeigen in entgegengesetzte Richtungen).  

Zur Veranschaulichung lassen sich Vektoren mithilfe spezieller Verfahren wie der **PCA (Principal Component Analysis)** auf zwei Dimensionen reduzieren, ohne dass ihre Ähnlichkeitsbeziehungen verloren gehen — wie genau PCA funktioniert, ist für uns irrelevant.

```python
from transformers import AutoModel
from sklearn.decomposition import PCA
import numpy as np
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-german-dbmdz-uncased")


# Einstellen die Ausgabeoptionen von numpy
np.set_printoptions(precision=18, suppress=True)

# Modell laden
model = AutoModel.from_pretrained("bert-base-german-dbmdz-uncased")

# Zwei Token-IDs hier einfügen
token_id1 = 4000
token_id2 = 3000

# Token-Wortstücke aus IDs holen
token1 = tokenizer.convert_ids_to_tokens(token_id1)
token2 = tokenizer.convert_ids_to_tokens(token_id2)

# Embeddings extrahieren und zu numpy konvertieren
vec1 = model.embeddings.word_embeddings.weight[token_id1].detach().cpu().numpy()
vec2 = model.embeddings.word_embeddings.weight[token_id2].detach().cpu().numpy()

# Stapeln & PCA auf 2D
pca = PCA(n_components=2, whiten=False)
vectors_2d = pca.fit_transform([vec1, vec2])

# Ausgabe
print("2D-Vektor 1:", vectors_2d[0], f"({token1})")
print("2D-Vektor 2:", vectors_2d[1], f"({token2})")
```

# **Fragebogen für eine Untersuchung**

```python
import ipywidgets as widgets
from IPython.display import display, HTML

# Frage und Antworten
frage = "Welche dieser Tiere sind Säugetiere?"
antworten = ["Krokodil", "Hund", "Frosch", "Katze"]
korrekte_indices = [1, 3]  # Hund und Katze

# Checkboxen und Labels vorbereiten
checkboxen = [widgets.Checkbox(value=False) for _ in antworten]
labels = [widgets.Label(value=antworten[i]) for i in range(len(antworten))]

# HBox-Zeilen kombinieren
zeilen = [widgets.HBox([checkboxen[i], labels[i]]) for i in range(len(antworten))]

# Lösung anzeigen
def zeige_loesung(b):
    for i in range(len(antworten)):
        if i in korrekte_indices:
            labels[i].value = f"✅ **{antworten[i]}**"
        else:
            labels[i].value = f"{antworten[i]}"
    b.disabled = True

# Button erstellen
loesung_button = widgets.Button(description="Lösung zeigen", button_style="info")
loesung_button.on_click(zeige_loesung)

# Anzeige
display(HTML(f"<h3>{frage}</h3>"))
display(widgets.VBox(zeilen))
display(loesung_button)
```
